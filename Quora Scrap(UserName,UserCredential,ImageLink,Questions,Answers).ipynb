```python
def quora_scrap(input_id):
    import time
    from selenium.webdriver.common.keys import Keys
    from selenium import webdriver

    import urllib
    from bs4 import BeautifulSoup

    driver = webdriver.Chrome(executable_path = 'G:\chrome_downloads\chromedriver_win32\chromedriver.exe')


    #load website to scrape
    driver.get("https://www.quora.com/profile/{}/answers".format(input_id))
    html1=driver.page_source
    # class for question
    #_base___StyledSpan-sc-1kk7hpy-0 iXGUkO rendered_qtext

    # class for answer
    #"Box-sc-9env3-0 Text-jjrgda-0 cYScOC">

    #Scroll the webpage
    ScrollNumber=5 #max scrolls-> it's value depends on count of questions-answers rendering through particular profile page. 

    for i in range(1, ScrollNumber):
        

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)

    # links generation for each answer according to given profile
    elem=driver.find_elements_by_class_name('ui_qtext_more_link')
    links=[]
    for lks in elem:
        links.append(lks.get_attribute('href'))


    #profile_photo_img

    img=driver.find_element_by_class_name("profile_photo_img").get_attribute('src')
    print('ImageLink :',img)

    # user name
    name=driver.find_element_by_class_name("user").text
    print('Name:',name)

    # user credential
    credential = driver.find_element_by_class_name('UserCredential').text
    print('Credential:',credential)
    
    soup=BeautifulSoup(html1)
    stats=[]
    stats_of_profile=soup.findAll('span',attrs={'class':'list_count'})
    for up in stats_of_profile:
        d=up.contents[0]
        stats.append(d)


    # followers
    print('FOLLOWERS',stats[4])

    # Answers
    print('ANSWERS',stats[0])

    # Questions
    print('QUESTIONS',stats[1])

    # Shares
    print('SHARES',stats[2])

    # Posts
    print('POSTS',stats[3])

    # Following
    print('FOLLOWING',stats[5])

    thumbnail = driver.find_elements_by_tag_name('iframe')
    tbn=[]
    for tb in thumbnail:
        tbn.append(tb.get_attribute('allowfullscreen id'))

    print('THUMBNAIL',tbn)
    q_a=dict()
    count=0
    print("QUESTIONS AND ANSWERS")
    for lks in links:
        driver.get(lks)
        ScrollNumber = 3
        for i in range(1, ScrollNumber):

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        html=driver.page_source
        html=html.encode()
        soup = BeautifulSoup(html)
        upvotes = driver.find_elements_by_class_name('Box-sc-9env3-0 cyUdtN')

        view = []
        views=soup.findAll('span',attrs={'class':'Box-sc-9env3-0 AbstractSeparatedItems___StyledBox-sc-46kfvf-0 chidNU'})
        for up in views:
            d=up.contents[0]
            view.append(d)
        time=[]
        timestamp=soup.findAll('a',attrs={'class':'Box-sc-9env3-0 eZSsnB'})
        for up in timestamp:
            d=up.contents[0]
            time.append(d)


        c1=[]
        for hit in soup.findAll(attrs={'class':'_base___StyledSpan-sc-1kk7hpy-0 iXGUkO rendered_qtext'}):
            d1=hit.contents[0]
            c1.append(d1)
        c2=[]
        for hit in soup.findAll('span',{'style':'font-weight: normal; font-style: normal;'}):
            d2=hit.contents[0]
            c2.append(d2)
        count+=1
        if count==5:
            break
        print('\nQUESTION-{}\n '.format(count),c1[0],'\nANSWER-{}\n '.format(count),' '.join(c2),'\nTIME ',time[1])
        if len(view)==4:
            print('\nVIEWS ',view[2])
        else:
            print('\nVIEWS ',view[1])
        print('\nUPVOTES ',upvotes)
        q_a[c1[0]]=c2

    print("Question-Answer-Dictionary",q_a)

# At the end you will get name,credential,imagelink and dictionary of question-answers...
################################################################################################################################


############################################   PENDING WORK   ##################################################################

# VIDEO_LINKS extraction from each answer
# If we can do this also then it will be scrapping of whole quora...


```
